Whiche Top 3 Color by Sales ?

Whiche Least 3 Colors by Sales ?

```python
import pandas as pd
import matplotlib.pyplot as plt

least_3 = df.groupby("Color")["Purchase Amount (USD)"].sum().reset_index()
least_3 = least_3.sort_values(by="Purchase Amount (USD)", ascending=True).tail(3)

plt.figure(figsize=(6,4))
plt.bar(least_3["Color"], least_3["Purchase Amount (USD)"])

plt.title("Least 3 Colors by Sales")
plt.xlabel("Color")
plt.ylabel("Purchase Amount (USD)")

plt.show()
```
